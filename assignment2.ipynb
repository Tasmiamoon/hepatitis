```python
df=df.drop(['Agebin','billirubibN','protime bin'],axis=1)
```

```python
df=pd.get_dummies(df,columns=["SEX","ANTIVIRALS","FATIGUE","MALAISE","ANOREXIA","LIVER BIG","LIVER FIRM",
                              "SPLEEN PALPABLE","SPIDERS","ASCITES","VARICES"])
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler=StandardScaler()
columns_to_scale=["AGE","AGE","SGOT","SGOT","ALBUMIN","PROTIME"]
df[columns_to_scale]=standardscaler.fit_transform(df[columns_to_scale])
```

```python
#df["CLASS"].replace({1: 0, 2: 1}, inplace=True)
```

```python
df.head()
```

```python
#splitting the feature data and target data
X=df.drop(["CLASS"],axis=1)
y=df["CLASS"]
print(X.shape,y.shape)
```

```python
sns.pairplot(df,hue="CLASS")
```

**50% training and 50% testing**

```python
#split the data again 50%training and 50% testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.5,random_state=1)
```

```python
#Using Random tree classifier with default parameter 
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt_trained=dt.fit(X_train,y_train)
y_pred=dt_trained.predict(X_test)
dt
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
  

results = confusion_matrix(y_test,y_pred) 
  
print 'Confusion Matrix :'
print(results) 
print 'Accuracy Score :',accuracy_score(y_test,y_pred) 
print 'Report : '
print classification_report(y_test,y_pred) 
```

```python
%matplotlib inline
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO  
from IPython.display import Image  
import pydot
import pydotplus

dot_data = StringIO()
f=df.columns[1:]
features = [u'{}'.format(c) for c in f]
export_graphviz(dt, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names=features,class_names=['1','2'])
graph = pydot.graph_from_dot_data(dot_data.getvalue())[0]  
Image(graph.create_png())  
```

```python
#using neural network model with hidden layer sizes(3,3)
from sklearn.neural_network import MLPClassifier
MLP=MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(3, 3), random_state=1)
MLP_trained=MLP.fit(X_train,y_train)
y_pred=MLP_trained.predict(X_test)
MLP
```

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 
  

results = confusion_matrix(y_test,y_pred) 
  
print 'Confusion Matrix :'
print(results) 
print 'Accuracy Score :',accuracy_score(y_test,y_pred) 
print 'Report : '
print classification_report(y_test,y_pred) 
```
